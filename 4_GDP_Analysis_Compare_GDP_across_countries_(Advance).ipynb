```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import plotly.offline as pyo # for html file
```

```python
df = pd.read_csv('/content/sample_data/gdp.csv')
```

```python
final_data = []
for country_name in df['Country Name'].unique():
  df_pr = df[df['Country Name'] == country_name]

  data = df_pr.values

  gdp_change = [0]

  for i in range(1,len(data)):

    prev = data[i-1][3]
    cur = data[i][3]
    gdp_change.append(round(((cur - prev)/prev)*100,2))
  df_pr = df_pr.assign(gdp_changes = gdp_change)
  final_data.append(df_pr)
df = pd.concat(final_data, axis = 0)
```

```python
c1 = df[df['Country Name'] == 'Arab World']
c2 = df[df['Country Name'] == 'India']
df_pr = pd.concat([c1,c2], axis = 0)
fig = px.line(df_pr,x = 'Year', y = 'Value', color = ('Country Name'), title = 'Analysis between India and Arab World')
fig
```
